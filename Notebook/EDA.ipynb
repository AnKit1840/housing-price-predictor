# Exploratory Data Analysis (EDA)

Here we are exploring the housing dataset to understand 
the structure, distributions, and relationships between variables before
training machine learning models.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
housing = pd.read_csv("housing.csv")
housing.head()
```

```python
housing.value_counts()
```

```python
housing["ocean_proximity"].value_counts()
```

```python
housing.info()
```

```python
housing.describe()
```

```python
housing.hist(bins=50, figsize=(12,8))
plt.show()
```

```python
housing.plot(
    kind="scatter",
    x="longitude",
    y="latitude",
    alpha=0.4
)
plt.title("Housing Locations")
plt.show()
```

```python
housing.plot(
    kind="scatter",
    x="longitude",
    y="latitude",
    alpha=0.4,
    s=housing["population"]/100,
    label="population",
    figsize=(10,7),
    c="median_house_value",
    cmap="jet",
    colorbar=True
)
plt.legend()
plt.show()
```

```python
corr_matrix = housing.corr(numeric_only=True)

corr_matrix["median_house_value"].sort_values(ascending=False)
```
